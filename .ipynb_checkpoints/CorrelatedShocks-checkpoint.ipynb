```julia
vec2mat=v->reshape(v,length(v),1)
```

```julia
T=1000;
```

```julia
g=0.030;
g_u=0.015;
```

```julia
struct OLSResult
    β::Vector{Float64}
    standard_errors::Vector{Float64}
    R²::Float64
end

function compute_ols(X::Matrix{Float64}, y::Vector{Float64})::OLSResult
    # Ensure a constant is included
    X = hcat(ones(size(X, 1)), X)

    # Estimate coefficients
    β = inv(X' * X) * X' * y

    # Calculate residuals
    residuals = y - X * β

    # Calculate R²
    TSS = sum((y .- mean(y)).^2)
    RSS = sum(residuals.^2)
    R² = 1 - (RSS / TSS)

    # Calculate standard errors
    s2 = RSS / (size(X, 1) - size(X, 2))
    var_cov_matrix = s2 * inv(X' * X)
    standard_errors = sqrt.(diag(var_cov_matrix))

    return OLSResult(β, standard_errors, R²)
end
```

```julia
err=randn(T);
```

```julia
x=rand(Normal(4,1),1000);
```

```julia
corr_coef=0.8;
```

```julia
eps=randn(T);
```

```julia
for i=2:T
    err[i]=corr_coef*err[i-1]+eps[i];
end
```

```julia
err=sqrt(1-corr_coef^2)*err
```

```julia
pstar=2;
```

```julia
p=pstar*ones(T,1);
```

```julia
H=fld(T,2);
```

```julia
V=zeros(T,1);
```

```julia
u_n=6*ones(T,1);
```

```julia
u=zeros(T,1);
```

```julia
for t=2:T
    u[t]=u_n[t]-V[t-1];
end
```

```julia
theta=-0.5;
```

```julia
for t=H:T
    p[t]=p[t-1]-theta*(u[t]-u_n[t])+err[t];
end
```

```julia
p_c=copy(p);
```

```julia
for t=H:T
    p_c[t]=p_c[t-1]-theta*(u[t]-u_n[t])+eps[t];
end
```

```julia
fix=800;
```

```julia
ustar_hat=copy(u);
```

```julia
for t=2:T
    ustar_hat[t]=ustar_hat[t-1]+g_u*(u[t]-ustar_hat[t-1]-V[t-1]);
end
```

```julia
pc_prior=compute_ols([p[2:fix-1] u[2:fix-1]-ustar_hat[2:fix-1,]-V[1:fix-2]],p[3:fix])
```

```julia
coefp=pc_prior.β
```

```julia
pc_prior_cf=compute_ols([p_c[2:fix-1] u[2:fix-1]-ustar_hat[2:fix-1,]],p_c[3:fix])
```

```julia
pc_prior_cf.β
```

```julia
plot([1:T],p)
```

```julia
plot([1:T],p_c)
```

```julia
res=p[3:fix]-[ones(fix-2) p[2:fix-1] u[2:fix-1]-ustar_hat[2:fix-1,]-V[1:fix-2]]*coefp
```

```julia
xcp=res[2:end-1];
```

```julia
cp_est=compute_ols(vec2mat(xcp),res[3:end])
```

```julia
coefcp=cp_est.β
```

```julia
cp_est.standard_errors
```

```julia
c_hat=zeros(T,1);
alpha_hat=zeros(T,1);
theta_hat=zeros(T,1);
psi_hat=zeros(T,1);
```

```julia
for t=fix+1:T
    xnairu=1;

    xxnairu=xnairu'*xnairu;
    ustar_hat[t]=ustar_hat[t-1]+g_u*inv(xxnairu)*xnairu*(u[t]-ustar_hat[t-1]);
    
    xp=[1 p[t-1] u[t-1]-ustar_hat[t-1]];
    xxp=xp'*xp;
    coefp=coefp+g*inv(xxp)*xp*(p[t]-xp*coefp');
    
    c_hat[t]=coefp[1];
    alpha_hat[t]=coefp[2];
    theta_hat=coefp[3];
    
    psi_hat[t]=psi_hat[t-1]+g*inv(xxcp)*xcp*(cp_hat[t]-psi[t-1]*cp_hat[t-1])
        
end
```
